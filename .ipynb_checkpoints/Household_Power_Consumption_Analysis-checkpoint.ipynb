## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline
print('Libraries imported!')
```

---
# 1. Data Exploration

```python
# Load data
df_raw = pd.read_csv('household_power_consumption.txt', sep=';', low_memory=False, na_values=['?', ''])
print(f'Raw data: {df_raw.shape[0]:,} rows, {df_raw.shape[1]} columns')
df_raw.head()
```

```python
# Create datetime index
df_raw['Datetime'] = pd.to_datetime(df_raw['Date'] + ' ' + df_raw['Time'], format='%d/%m/%Y %H:%M:%S')
df_raw.set_index('Datetime', inplace=True)
df_raw.drop(['Date', 'Time'], axis=1, inplace=True)
for col in df_raw.columns:
    df_raw[col] = pd.to_numeric(df_raw[col], errors='coerce')
print(f'Date range: {df_raw.index.min()} to {df_raw.index.max()}')
```

```python
# Aggregate to daily
df_daily = df_raw.resample('D').mean()
df = df_daily.interpolate(method='linear').fillna(method='bfill').fillna(method='ffill')
ts = df['Global_active_power'].copy()
print(f'Daily data: {len(ts)} days')
print(f'Missing values: {ts.isnull().sum()}')
```

```python
# Visualize
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(ts.index, ts.values, linewidth=0.8)
ax.set_title('Daily Global Active Power (2006-2010)', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Power (kW)')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].hist(ts, bins=50, edgecolor='black', alpha=0.7)
axes[0].axvline(ts.mean(), color='red', linestyle='--', label=f'Mean: {ts.mean():.2f}')
axes[0].set_title('Distribution', fontweight='bold')
axes[0].set_xlabel('Power (kW)')
axes[0].legend()

df['Month'] = df.index.month
monthly = [df[df['Month']==m]['Global_active_power'].values for m in range(1,13)]
axes[1].boxplot(monthly, labels=['J','F','M','A','M','J','J','A','S','O','N','D'])
axes[1].set_title('Monthly Pattern', fontweight='bold')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Power (kW)')
plt.tight_layout()
plt.show()
```

```python
# ADF Test
def adf_test(series, title):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'\nADF Test: {title}')
    print(f'Statistic: {result[0]:.4f}, p-value: {result[1]:.4f}')
    print('STATIONARY' if result[1] <= 0.05 else 'NON-STATIONARY')
    return result

adf_test(ts, 'Original')
adf_test(ts.diff().dropna(), 'Differenced')
```

```python
# ACF/PACF
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
plot_acf(ts, lags=40, ax=axes[0,0])
axes[0,0].set_title('ACF - Original')
plot_pacf(ts, lags=40, ax=axes[0,1], method='ywm')
axes[0,1].set_title('PACF - Original')
plot_acf(ts.diff().dropna(), lags=40, ax=axes[1,0])
axes[1,0].set_title('ACF - Differenced')
plot_pacf(ts.diff().dropna(), lags=40, ax=axes[1,1], method='ywm')
axes[1,1].set_title('PACF - Differenced')
plt.tight_layout()
plt.show()
```

---
# 2. Decomposition

```python
# Classical decomposition
decomp = seasonal_decompose(ts, model='additive', period=7)
fig, axes = plt.subplots(4, 1, figsize=(14, 10))
axes[0].plot(ts); axes[0].set_title('Original')
axes[1].plot(decomp.trend); axes[1].set_title('Trend')
axes[2].plot(decomp.seasonal); axes[2].set_title('Seasonal')
axes[3].plot(decomp.resid); axes[3].set_title('Residual')
plt.suptitle('Classical Decomposition (period=7)', fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# STL decomposition
stl = STL(ts, period=7, robust=True).fit()
fig, axes = plt.subplots(4, 1, figsize=(14, 10))
axes[0].plot(ts); axes[0].set_title('Original')
axes[1].plot(stl.trend); axes[1].set_title('Trend (STL)')
axes[2].plot(stl.seasonal); axes[2].set_title('Seasonal (STL)')
axes[3].plot(stl.resid); axes[3].set_title('Residual (STL)')
plt.suptitle('STL Decomposition (period=7)', fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---
# 3. Initial Forecasting

```python
# Train-test split
train_size = int(len(ts) * 0.8)
train, test = ts[:train_size], ts[train_size:]
print(f'Train: {len(train)}, Test: {len(test)}')
```

```python
# ARIMA
print('Fitting ARIMA(1,1,1)...')
arima = ARIMA(train, order=(1,1,1)).fit()
arima_fc = arima.forecast(len(test))
arima_fc.index = test.index
arima_rmse = np.sqrt(mean_squared_error(test, arima_fc))
arima_mae = mean_absolute_error(test, arima_fc)
print(f'ARIMA - RMSE: {arima_rmse:.4f}, MAE: {arima_mae:.4f}')
```

```python
# SARIMA
print('Fitting SARIMA(1,1,1)(1,1,1,7)...')
sarima = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,7)).fit(disp=False)
sarima_fc = sarima.forecast(len(test))
sarima_fc.index = test.index
sarima_rmse = np.sqrt(mean_squared_error(test, sarima_fc))
sarima_mae = mean_absolute_error(test, sarima_fc)
print(f'SARIMA - RMSE: {sarima_rmse:.4f}, MAE: {sarima_mae:.4f}')
```

```python
# Prophet
print('Fitting Prophet...')
prophet_train = pd.DataFrame({'ds': train.index, 'y': train.values})
prophet = Prophet(daily_seasonality=False, weekly_seasonality=True, yearly_seasonality=True)
prophet.fit(prophet_train)
future = prophet.make_future_dataframe(periods=len(test), freq='D')
pred = prophet.predict(future)
prophet_fc = pred[pred['ds'].isin(test.index)].set_index('ds')['yhat'].reindex(test.index).interpolate()
prophet_rmse = np.sqrt(mean_squared_error(test, prophet_fc))
prophet_mae = mean_absolute_error(test, prophet_fc)
print(f'Prophet - RMSE: {prophet_rmse:.4f}, MAE: {prophet_mae:.4f}')
```

```python
# Model comparison
print('\n' + '='*50)
print('MODEL COMPARISON')
print('='*50)
comp = pd.DataFrame({
    'Model': ['ARIMA', 'SARIMA', 'Prophet'],
    'RMSE': [arima_rmse, sarima_rmse, prophet_rmse],
    'MAE': [arima_mae, sarima_mae, prophet_mae]
})
print(comp.round(4).to_string(index=False))
print(f'\nBest: {comp.loc[comp["RMSE"].idxmin(), "Model"]}')
```

```python
# Plot forecasts
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(test.index[:60], test.values[:60], label='Actual', color='black', linewidth=2)
ax.plot(test.index[:60], arima_fc.values[:60], label='ARIMA', linestyle='--')
ax.plot(test.index[:60], sarima_fc.values[:60], label='SARIMA', linestyle='--')
ax.plot(test.index[:60], prophet_fc.values[:60], label='Prophet', linestyle='--')
ax.set_title('Forecast Comparison (60 days)', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---
# 4. Anomaly Detection

```python
# Z-Score
z_scores = (ts - ts.mean()) / ts.std()
zscore_anom = np.abs(z_scores) > 3
print(f'Z-Score anomalies: {zscore_anom.sum()}')
```

```python
# ARIMA Residuals
arima_full = ARIMA(ts, order=(1,1,1)).fit()
resid = arima_full.resid
resid_z = (resid - resid.mean()) / resid.std()
arima_anom = np.abs(resid_z) > 3
print(f'ARIMA anomalies: {arima_anom.sum()}')
```

```python
# Isolation Forest
features = pd.DataFrame(index=ts.index)
features['Power'] = ts
features['Roll7'] = ts.rolling(7).mean()
features['Roll30'] = ts.rolling(30).mean()
features['Diff'] = ts.diff()
features = features.dropna()

iso = IsolationForest(contamination=0.02, random_state=42)
iso_pred = iso.fit_predict(StandardScaler().fit_transform(features))
iso_anom = iso_pred == -1
print(f'Isolation Forest anomalies: {iso_anom.sum()}')
```

```python
# Consensus
idx = features.index
summary = pd.DataFrame(index=idx)
summary['Power'] = features['Power']
summary['Z'] = zscore_anom.reindex(idx).fillna(False)
summary['A'] = arima_anom.reindex(idx).fillna(False)
summary['I'] = iso_anom
summary['Count'] = summary['Z'].astype(int) + summary['A'].astype(int) + summary['I'].astype(int)
summary['Consensus'] = summary['Count'] >= 2
consensus_dates = summary[summary['Consensus']].index

print(f'\nConsensus anomalies (2+): {len(consensus_dates)}')
```

```python
# Visualize
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(ts.index, ts.values, linewidth=0.7, alpha=0.7)
ax.scatter(consensus_dates, ts.loc[consensus_dates], color='red', s=40, label=f'Anomalies ({len(consensus_dates)})')
ax.set_title('Power Consumption with Consensus Anomalies', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---
# 5. Re-Forecasting After Cleaning

```python
# Clean data
ts_clean = ts.copy()
ts_clean.loc[ts_clean.index.isin(consensus_dates)] = np.nan
ts_clean = ts_clean.interpolate(method='linear')
print(f'Cleaned {len(consensus_dates)} anomalies')
```

```python
# Re-fit models
train_c, test_c = ts_clean[:train_size], ts_clean[train_size:]

# ARIMA
arima_c = ARIMA(train_c, order=(1,1,1)).fit()
arima_c_fc = arima_c.forecast(len(test))
arima_c_fc.index = test.index
arima_c_rmse = np.sqrt(mean_squared_error(test, arima_c_fc))

# SARIMA
sarima_c = SARIMAX(train_c, order=(1,1,1), seasonal_order=(1,1,1,7)).fit(disp=False)
sarima_c_fc = sarima_c.forecast(len(test))
sarima_c_fc.index = test.index
sarima_c_rmse = np.sqrt(mean_squared_error(test, sarima_c_fc))

# Prophet
prophet_c_train = pd.DataFrame({'ds': train_c.index, 'y': train_c.values})
prophet_c = Prophet(daily_seasonality=False, weekly_seasonality=True, yearly_seasonality=True)
prophet_c.fit(prophet_c_train)
future_c = prophet_c.make_future_dataframe(periods=len(test), freq='D')
pred_c = prophet_c.predict(future_c)
prophet_c_fc = pred_c[pred_c['ds'].isin(test.index)].set_index('ds')['yhat'].reindex(test.index).interpolate()
prophet_c_rmse = np.sqrt(mean_squared_error(test, prophet_c_fc))

print('Models re-fitted on cleaned data!')
```

```python
# Comparison
print('\n' + '='*60)
print('BEFORE vs AFTER CLEANING')
print('='*60)
arima_imp = (arima_rmse - arima_c_rmse) / arima_rmse * 100
sarima_imp = (sarima_rmse - sarima_c_rmse) / sarima_rmse * 100
prophet_imp = (prophet_rmse - prophet_c_rmse) / prophet_rmse * 100

print(f'ARIMA:   {arima_rmse:.4f} -> {arima_c_rmse:.4f} ({arima_imp:+.2f}%)')
print(f'SARIMA:  {sarima_rmse:.4f} -> {sarima_c_rmse:.4f} ({sarima_imp:+.2f}%)')
print(f'Prophet: {prophet_rmse:.4f} -> {prophet_c_rmse:.4f} ({prophet_imp:+.2f}%)')
```

```python
# Bar chart
fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(3)
w = 0.35
ax.bar(x-w/2, [arima_rmse, sarima_rmse, prophet_rmse], w, label='Original', color='#d62728')
ax.bar(x+w/2, [arima_c_rmse, sarima_c_rmse, prophet_c_rmse], w, label='Cleaned', color='#2ca02c')
ax.set_xticks(x)
ax.set_xticklabels(['ARIMA', 'SARIMA', 'Prophet'])
ax.set_ylabel('RMSE')
ax.set_title('Model Performance: Before vs After Cleaning', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

---
# 6. Conclusion

```python
all_rmse = {'ARIMA Orig': arima_rmse, 'ARIMA Clean': arima_c_rmse,
            'SARIMA Orig': sarima_rmse, 'SARIMA Clean': sarima_c_rmse,
            'Prophet Orig': prophet_rmse, 'Prophet Clean': prophet_c_rmse}
best = min(all_rmse, key=all_rmse.get)

print(f'''
{'='*60}
FINAL SUMMARY
{'='*60}

DATASET: UCI Household Power Consumption
PERIOD: {ts.index.min().date()} to {ts.index.max().date()}
DAYS: {len(ts)}

ANOMALIES DETECTED: {len(consensus_dates)} (consensus)

MODEL PERFORMANCE (RMSE):
         Original   Cleaned    Change
ARIMA:   {arima_rmse:.4f}     {arima_c_rmse:.4f}     {arima_imp:+.2f}%
SARIMA:  {sarima_rmse:.4f}     {sarima_c_rmse:.4f}     {sarima_imp:+.2f}%
Prophet: {prophet_rmse:.4f}     {prophet_c_rmse:.4f}     {prophet_imp:+.2f}%

BEST MODEL: {best} (RMSE: {all_rmse[best]:.4f})

KEY FINDINGS:
1. Clear weekly seasonality in power consumption
2. SARIMA captures seasonal patterns effectively
3. Anomaly cleaning can improve forecast accuracy
4. Energy data is more predictable than stock prices
{'='*60}
''')
```

```python
# Save results
results = pd.DataFrame({
    'Model': ['ARIMA', 'SARIMA', 'Prophet'],
    'RMSE_Original': [arima_rmse, sarima_rmse, prophet_rmse],
    'RMSE_Cleaned': [arima_c_rmse, sarima_c_rmse, prophet_c_rmse],
    'Improvement_%': [arima_imp, sarima_imp, prophet_imp]
})
results.to_csv('model_results.csv', index=False)
print('Results saved!')
results.round(4)
```
